# 1. Regression Analysis _(Phân tích hồi quy)_
* Dùng để dự báo, dự đoán một giá trị mà giá trị này là một giá trị số nằm trong một khoảng nào đó.
* Là một dạng của **predictive modelling technique** _(kĩ thuật mô hình tiên đoán)_, nhiệm vụ của nó là điều tra mối quan hệ giữa **biến phụ thuộc** _(dependent variable, target)_ và các **biến độc lập** _(independent variable, predictor)_.

  ![](../images/0.png)

* Mục tiêu là dùng để dự đoán giá trị số từ input.
* Lưu ý output luôn luôn là giá trị số trong một phạm vi nào đó.
* Regression là **supervised** _(học có giám sát)_.

  ![](../images/1.png)

* Ví dụ:
  * Ước tính giá nhà trung bình cho một khu vực.
  * Xác định nhu cầu cho một sản phẩm mới
  * Dự đoán mức sử dụng năng lượng, điện, nước...
  

## 1.1. Xây dựng và áp dụng model
* Bao gồm 2 giai đoạn:
  * Training phase _(huấn luyện)_
    * Điều chỉnh **tham số model** _(model parameter)_.
    * Sử dụng **dữ liệu huấn luyện** _(training data)_.
  * Testing phase _(kiểm chứng)_
    * Áp dụng **mô hình đã học** _(learned model)_.
    * Sử dụng **dữ liệu mới** _(new data)_.
* Nếu model ko sử dụng được thì phải xem các nguyên nhân dưới đây:
  * Do dữ liệu ít nên ko chính xác.
  * Do dữ liệu còn rác vì tiền xử lí dữ liệu ko tốt.
  * Model này quá đơn giản/phức tạp so với dữ liệu ta có.
  * ...
* Các bước thực hiện:
  ![](../images/2.png)
  * Chia dữ liệu thành training data và test data, training data chiếm khoảng 70 đến 80% của dữ liệu gốc và phần còn lại là test data.
  * Tiếp theo cần chọn thuật toán và sau đó xây dựng một model bằng thuật toán này.
  * Sau khi xây dựng model thành công, lấy tập dữ liệu test data để kiểm chứng chất lượng của model này.

## 1.2. Một số thuật toán Regression
* Một vài thuật toán phổ biến như sau:
  * Linear Regression
  * Polynomial Regression
  * Regression Trees & Random Forest
  * KNN
  * SVM
  * ...
* Mỗi bộ dữ liệu có một đặc trưng riêng do đó phù hợp vs 1 số thuật toán riêng của nó, tức cần có một chiến lược chọn ra một thuật toán dành cho bộ dữ liệu này.

# 2. Linear Regression
* Là một thuật toán thuộc nhóm Supervised Learning dc sử dụng cho regression.
* Được sử dụng rộng rãi.
* Ví dụ:
  * Quản lí rũi ro hoặc chấm điểm tín dụng của ngân hàng, hệ thống sẽ dựa trên các thông tin khách hàng về mức lương hiện tại, độ tuổi, công việc,... để quyết định xem khoản tiền, thời hạn vay cho một khách hàng.
  * Dự báo tài chính.
  * Dự đoán chi phí phần mềm, chi phí đảm bảo chất lượng phần mềm.

## 2.1. Ý tưởng thuật toán
* Là tìm dc **mối quan hệ tuyến tính** _(linear relationship)_ giữa output và input.
  ![](../images/3.png)
* Dependent variable là liên tục, còn các independent variable có thể là liên tục hoặc rời rạc.
* Mô hình dc cho là phù hợp bằng cách sử dụng thuật toán **least squares** để kiểm chứng.
* Ngoài ra có thể áp dụng kí thuật correlation trên thang đo từ $[-1,1]$ để kiểm tra mối tương quan giữa hai biến, nếu:
  * Correlation bằng 1 hoặc -1 thì perfect.
  * Trong khoảng $[0.7, 0.9]$ hoặc $[-0.9, -0.7]$ thì high.
  * Trong khoảng $[0.4, 0.6]$ hoặc $[-0.6, -0.4]$ thì weak.
  * Trong khoảng $[-0.3, 0.3]$ thì ko có mối tương quan.
    ![](../images/4.png)

## 2.2. Model
![](../images/5.png)
* Khi model được xây dựng xong thì nó là đường màu đỏ _(regression line)_ ở hình trên.

## 2.3. Phương trình của Linear Regression
![](../images/6.png)
* Cần đi tìm hai tham số $m$ _(slope))_ và $b$ _(y-intercept)_.
* Hai giá trị $m$ và $b$ được tìm ra bằng thuật toán **Gradient Descent**.
* Model dc huấn luyện bởi linear regression sẽ điều chỉnh model parameter cho phù hợp vs mẫu.

## 2.4. Phương pháp Least Squares Method
* Câu hỏi đặt ra là làm sao để ta biết dc regression line mà ta tìm dc liệu có phù hợp hay ko để ta chấp nhận nó hoặc tìm một regression line khác phù hợp hơn với tập dữ liệu. Lúc này ta sẽ áp dụng least squares method, phương pháp này sẽ tính tổng các số dư, với số dư là khoảng cách từ một điểm dữ liệu cụ thể đến đường regression line, nếu tổng các số dư càng nhỏ thì đường regression line này càng tốt.
  ![](../images/7.png)
* Có ba cách để tính ra gía trị least squares:
  ![](../images/8.png)

## 2.5. Áp dụng mô hình
* Cho $X$ vào model, model sẽ dự đoán ra giá trị $Y$ tương ứng
  ![](../images/9.png)

## 2.6. Phân loại thuật toán
* Bao gồm hai loại:
  * Simple Linear Regression: chỉ có một $X$ _(features)_ để dự đoán ra $Y$.
    ![](../images/10.png)
  * Multiple Linear Regression: có nhiều $X$ _(features)_ để dự doán ra $Y$.
    ![](../images/11.png)
* Dạng multiple thường dc sử dụng trong thực tế nhiều hơn so với simple.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_profiling as pp
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r'./data/location_rev.csv')
df.head()
```

### Huấn luyện model với traning data 

```python
X1 = df[['median_income']]
y1 = df['revenue']

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, random_state=42)
model = LinearRegression()
model.fit(X1_train, y1_train)
```

### Kiểm tra model với testing data

```python
y1_pred = model.predict(X1_test)
intercept = model.intercept_ # b
slope = model.coef_[0]

(intercept, slope)
```

### Tính $R^2$

```python
model.score(X1, y1)
```

> Mô hình này chỉ

```python
plt.figure(figsize=(5, 5))
plt.scatter(y1_pred, y1_test)
plt.xlabel('Model predictions')
plt.ylabel('True value')
plt.plot([0, 100_000], [0, 100_000], 'k-', color='r')
```
